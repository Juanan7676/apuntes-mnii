```python
# Complex fractals from iterative methods
from PIL import Image

def draw_fractal(iterative_method, xa=-1.0, xb=1.0, ya=-1.0, yb=1.0):
    """Draw a fractal image defined from the number of iterations of
    a given method, starting for each complex point in a rectangular area.

    * Args: 
        iterative_method: function returning the number of iterations 
            of a given method starting from a complex point
    * Returns: 
        fractal image
    """

    imgx = 512
    imgy = 512
    image = Image.new("RGB", (imgx, imgy))

    # draw the fractal
    for y in range(imgy):
    	zy = y * (yb - ya) / (imgy - 1) + ya # Transform [0...imgy) -> [ya...yb)
	for x in range(imgx):
            zx = x * (xb - xa) / (imgx - 1) + xa # Transform [0...imgx) -> [xa...xb)
	    z = complex(zx, zy)
	    iters = iterative_method(z) # Number of iterations when we start from z
	    image.putpixel((x, y), (iters % 4 * 64, iters % 16 * 16, iters % 8 * 32))
    return image
```

```python
# Function whose zero is approximated using Newthon method
def f(z): return z*z*z - 1

# Newthon method
def newton_iterative_method(z):
    """Returns the number of iterations for Newton method, starting from z"""
    maxIt = 20 # max iterations allowed
    h = 1e-6 # step size for numerical derivative
    eps = 1e-3 # max error allowed

    for i in range(maxIt):
	# complex numerical derivative
	dz = (f(z + complex(h, h)) - f(z)) / complex(h, h)
	z0 = z - f(z) / dz # Newton iteration
	if abs(z0 - z) < eps: # stop when close enough to any root
	    break
	z = z0
    return i 

image = draw_fractal(newton_iterative_method)

imshow(image)
```

```python
def mandelbrot_iterative_method(c): 
    """Returns the number of iterations for which $z_n$ 'escapes' from a given ball,
    whith $z_{n+1}=z_n^2+c$, starting from a complex c""" 
    maxIt = 20 # max iterations allowed 
    R = 1e+3 # max size of $z$ allowed
    z = complex(0.0, 0.0)
    for i in range(maxIt):
	z0 = z*z + c # Mandelbrot iteration
	if abs(z0) > R: # stop when z0 is big 
	    break
	z = z0
    return i 

image = draw_fractal(mandelbrot_iterative_method, 
		     xa=-2.5, xb=1.5, ya=-1.5, yb=1.5)

imshow(image)
```
