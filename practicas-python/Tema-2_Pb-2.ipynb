# Tema 2. Problema 2

Consideremos en el intervalo $[-1,1]$ la función
\begin{equation*}
     f(x) = \frac{1}{1+25x^2}
\end{equation*}
junto a un soporte de $n+1$ puntos equiespaciados en $[-1,1]$.

Representar gráficamente la función junto al polinomio de interpolación de
Lagrange, $p_n(x)$, para $n=\{ 4, 8, 12, 16, 20 \}$. Comprobar gráficamente cómo el error aumenta con $n$.

```python
def f(x): return 1./(1+25*x*x)

# 1. Gráfica de f
a, b = -1, 1
x = linspace(a, b, 100)
axhline(c='k'); axvline(c='k') # Ejes Horizontal y Vertical
plot(x, f(x), "-", lw=5, c='midnightblue', label="$f(x)$")

# 2. Gráficas de polinomios de interpol.
from scipy.interpolate import lagrange
lista_n = [4, 8, 12]#, 16, 20] 
lista_colores = ["green", "red", "pink", "orange", "gold"]

for i in range( len(lista_n) ):
    n = lista_n[i]
    nodos = linspace(a, b, n+1)
    p = lagrange(nodos, f(nodos))
    c = lista_colores[i]
    plot(x, p(x), "-",color=c, lw=2, label="$p_{" + str(i) + "}(x)$")
grid()
legend()
```
