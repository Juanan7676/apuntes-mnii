```python
imagen = imread('UCA-CASEM.png') 
imshow(imagen)

xlist = [0,    60, 120, 180, 300, 400, 470, 540, 600]
ylist = [100, 106, 112, 115,  15, 116, 112, 106, 100]
plot(xlist, ylist, 'o', c='red')

from scipy.interpolate import lagrange
p_lagrange = lagrange(xlist, ylist)

xx = linspace(0,600,200)
plot(xx, p_lagrange(xx), c='orange')

from scipy.interpolate import interp1d
s_cubico = interp1d(xlist, ylist, kind='cubic')
plot(xx, s_cubico(xx), c='yellow')
```
