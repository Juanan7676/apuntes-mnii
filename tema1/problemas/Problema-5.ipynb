```python
def f(x): return 2*exp(x) - (x+0.5)**2
def df(x): return 2*exp(x) - 2*(x+0.5)

x = linspace(-2,-0)
plot(x, f(x),linewidth=3)
grid()
```

```python
def newton(f, df, x0, tol):
    """Calcula una raíz de $f(x)=0$ mediante el método de Newton, a partir de x0"""
    
    max_iters = 100
    iter = 0
    while iter<max_iters:
        x1 = x0-f(x0)/df(x0) # Punto fijo en Newton
        if abs(x1-x0) < tol:
            return x1
        x0 = x1
        iter = iter+1
    print "Error de convergencia"
    
x = newton(f, df, -2, 1e-8)
print x
```
